```python
class Archivo:
    def __init__(self, nombre):
        try:
            self.f = open(nombre, 'r') #Se crea un atributo de archivo
            self.nombre = nombre #El nombre de archivo
        except:
            print("No se puede abrir", nombre)
            exit()

    def muestra(self):
        i = 1
        for linea in self.f:
            print("{:3}:{}".format(i, linea)) #formato de salida
            i += 1
        self.f.seek(0) #Colocar el apuntador a la posicion inicial

    def cuentaVocales(self):
        def vocales(s):
            contador = 0
            for i in range(len(s)):
                if s[i] in set("aeiouáéíóú"): #comparara si el caracter existe en nuestro conjunto de vocales
                    contador += 1 #Si existe, suma 1 al contador
            return contador
        contador = 0
        for linea in self.f: #Linea por linea
            contador += vocales(linea) #Nuestro contador se va acumulando
        self.f.seek(0)#Pone el apuntador a posicion inicial
        return contador

    def cuentaConsonantes(self):
        def consonantes(s):
            contador = 0
            for i in range(len(s)):
                if s[i].lower() in set("bcdfghjklmnñpqrstvwxyz"): ##Verifica si existe en nueustro conjunto de consonantes
                    contador += 1
            return contador
        contador = 0
        for linea in self.f: #Linea por linea
            contador += consonantes(linea) #Nuestro contador se va acumulando
        self.f.seek(0)#Pone el apuntador a posicion inicial
        return contador

    def cuentaSignos(self):
        def Signos(s):
            contador = 0
            for i in range(len(s)):
                if s[i] in {"...", ",", ".", ";", '"'}: #Verifica que algun de los signos aparezca
                    contador += 1 #si si, suma un 1 al contador
            return contador
        contador = 0
        for linea in self.f:
            contador += Signos(linea) 
        self.f.seek(0)
        return contador
    
    def cuentaEspacios(self):
        def Espacios(s):
            contador = 0
            for i in range(len(s)):
                if s[i] in {" "}: #Verifica que haya un espacio
                    contador += 1
            return contador
        contador = 0
        for linea in self.f: #Va leyendo cada linea
            contador += Espacios(linea)
        self.f.seek(0)
        return contador
    
    
    def cuentaPalabras(self):
        def Palabras(s):
            contador = 0
            flg = 0
            i = 0
            while True:#ciclo infinito
                if i > len(s)-1:#esta condición lo detiene
                    return contador
                if s[i] in set("abcdefghijklmnñopqrstuvwxyzáéíóú"):#cuenta una palabra cuando encuentra una letra
                    contador += 1
                    while s[i] in set("abcdefghijklmnñopqrstuvwxyzáéíóú"):#sigue a traves de la palabra hasta encontrar algo diferente
                        i += 1
                else:
                    i += 1#si no es una letra solo sigue iterando
        contador = 0
        for linea in self.f:
            contador += Palabras(linea)
        self.f.seek(0)
        return contador
    
    def cuentaLineas(self):
        def Lineas(s):
            contador = 0
            for i in range(len(s)):
                if s[i]=='\n': #Verifica el salto de linea
                    contador +=1
            return contador
        contador = 0
        for linea in self.f: #Va leyendo cada linea
            contador += Lineas(linea)
        self.f.seek(0)
        return contador
    def cuentaMinus(self):
        def Minus(s):
            contador = 0
            for i in range(len(s)):
                if 96 < ord(s[i]) < 123: #Checa que el numero ASCII pertenezca a nuestro intervalo de enteros
                    contador += 1
            return contador
        contador = 0
        for linea in self.f: #Va leyendo cada linea
            contador += Minus(linea)
        self.f.seek(0)
        return contador
    
    def cuentaMayus(self):
        def Mayus(s):
            contador = 0
            for i in range(len(s)):
                if 64 < ord(s[i]) < 91: #Verifica que el numero ASCII de nuestro caracter pertenezca a nuestro intervalo de enteros
                    contador += 1
            return contador
        contador = 0
        for linea in self.f: #Va leyendo cada linea
            contador += Mayus(linea)
        self.f.seek(0)
        return contador
    
    def copiarCad(self, s):#solamente escribe la misma cadena a un archivo nuevo
        copia = open("copia", "w")
        copia.write(s)
        
    def enHex(self, s):
            for i in range(len(s)):
                print(hex(ord(s[i])), end=' ')#convierte cada caracter en int usando ord() y luego en hexadecimal usando hex()
    
nomb = input("Escribe el nombre de archivo:")
archivo = Archivo(nomb)
archivo.muestra()
print()
s = archivo.f.read()
archivo.copiarCad(s)

print("El numero de vocales:", archivo.cuentaVocales())
print("El numero de consonantes es:", archivo.cuentaConsonantes())
print("El numero de signos es:", archivo.cuentaSignos())
print('El numero de espacios es:', archivo.cuentaEspacios())
print('El numero de palabras es:', archivo.cuentaPalabras())
print('El numero de lineas es:', archivo.cuentaLineas())
print('El numero de mayusculas es:', archivo.cuentaMayus())
print('El numero de minusculas es:', archivo.cuentaMinus(),"(Las palabras acentuadas no son contadas)")
print('El archivo está copiado!\n')
cad = archivo.f.read()
aux = cad.upper()
print("Cadena en mayúsculas:\n", aux,'\n')
aux = cad.lower()
print("Cadena en minúsculas:\n", aux, '\n')
print("Cadena en hexadecimal:\n")
archivo.enHex(cad)
```
